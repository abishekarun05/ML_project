```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Video Games Sales (1980-2024).csv')
df
```

```python
plt.figure(figsize=(16, 8))
sns.countplot(data = df, x = 'genre', order = df['genre'].value_counts().index, palette = 'viridis')
plt.title('Number of Video Games by Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data = df, x = 'critic_score', bins = 30, kde = True, color = 'skyblue', edgecolor = 'black')
plt.title('Distribution of Critic Scores')
plt.xlabel('Critic Score')
plt.ylabel('Number of Games')
plt.tight_layout()
plt.show()
```

```python
genre_counts = df['genre'].value_counts()
top_n = 8
top_n_genres = genre_counts.head(top_n)
others = pd.Series({'Others': genre_counts.iloc[top_n:].sum()})
genre_pie = pd.concat([top_n_genres, others])
plt.figure(figsize=(10, 8))
colors = plt.cm.Set3.colors
explode = [0.05] + [0] * (len(genre_pie) - 1)
plt.pie(genre_pie, labels = genre_pie.index, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode, shadow=True)
plt.title('Distribution of Video Game Genres')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data = df, x = 'genre', y = 'critic_score', palette = 'Set2')
plt.title('Critic Scores Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Critic Score')
plt.xticks(rotation=45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data = df, x = 'critic_score', y = 'total_sales', hue = 'genre', alpha = 0.6, palette = 'tab10')
plt.title('Critic Score vs Total Sales by Genre')
plt.xlabel('Critic Score')
plt.ylabel('Total Sales (in millions)')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[numeric_cols].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', center=0, square=True, linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.violinplot(data = df, x = 'genre', y = 'critic_score', palette = 'Set2', inner = 'quartile')
plt.title("Critic Score Distribution by Genre (Violin Plot)")
plt.xlabel('Genre')
plt.ylabel('Critic Score')
plt.xticks(rotation=45, ha = 'right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data = df, x = 'critic_score', hue = 'genre', fill = True, common_norm = False, palette = 'Set2', alpha = 0.6)
plt.title('Kernel Density Estimation of Critic Scores by Genre')
plt.xlabel('Critic Score')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(df, col = 'genre', col_wrap = 4, height = 3, sharex = False, sharey = False, aspect = 1.2)
g.map(sns.histplot, 'critic_score', kde=True, bins = 20, color = 'skyblue', edgecolor = 'black')
g.set_axis_labels('Critic Score', 'Count')
g.set_titles('{col_name}')
g.fig.suptitle('Critic Score Distribution by Genre', fontsize = 16, y = 1.02)
plt.tight_layout()
plt.show()
```

```python
num_cols = ['critic_score', 'total_sales', 'na_sales', 'jp_sales']
sns.pairplot(df[num_cols + ['genre']], diag_kind='kde', plot_kws={'alpha':0.6, 's':40}, hue = 'genre', palette = 'tab10')
plt.suptitle('Pairplot of Numeric Variables by Genre', fontsize = 16, y = 1.02)
plt.tight_layout()
plt.show()
```

```python
sample_df = df.sample(n=3000, random_state=42)
plt.figure(figsize=(14, 7))
sns.swarmplot(data = sample_df, x = 'genre', y = 'critic_score', hue = 'genre', palette = 'Set2', size = 4, alpha = 0.85)
plt.title('Swarm Plot of Critic Scores by Genre (Sampled Data)', fontsize = 16, pad = 15)
plt.xlabel('Genre', fontsize = 12)
plt.ylabel('Critic Score', fontsize = 12)
plt.xticks(rotation=45, ha = 'right')
plt.tight_layout()
```
